```python
import pickle
history = pickle.load(open('mimic_ls', "rb"))
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
history
```

```python
loss = history['loss']
val_loss= history['val_loss']

accuracy = history['accuracy']
val_accuracy = history['val_accuracy']

precision = history['precision']
val_precision = history['val_precision']

recall = history['recall']
val_recall = history['val_recall']

```

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 15))

axs[0, 0].plot(range(0,len(loss)), loss, color = 'r')
axs[0, 0].plot(range(0,len(val_loss)), val_loss, color = 'b')
axs[0, 0].set_title('Funkcja loss')

axs[0, 1].plot(range(0,len(accuracy)), accuracy, color = 'r')
axs[0, 1].plot(range(0,len(val_accuracy)), val_accuracy, color = 'b')
axs[0, 1].set_title('Trafno (accuracy)')

axs[1, 0].plot(range(0,len(precision)), precision, color = 'r')
axs[1, 0].plot(range(0,len(val_precision)), val_precision, color = 'b')
axs[1, 0].set_title('Precyzja (Precision)')

axs[1, 1].plot(range(0,len(recall)), recall, color = 'r')
axs[1, 1].plot(range(0,len(val_recall)), val_recall, color = 'b')
axs[1, 1].set_title('Recall')

for row in axs: 
    for ax in row: 
        ax.grid(b=True, which='major', color='#666666', linestyle='-')
        ax.minorticks_on()  
        ax.grid(b=True, which='minor', color='#999999', linestyle='-',alpha=0.2) 

fig.legend(["Trening", "Walidacja"], loc='upper left', bbox_to_anchor=(0.1,0.1), ncol=2, bbox_transform=fig.transFigure)
```

```python
fig.savefig('sf_mimic.jpg', dpi=300, format='jpg', bbox_inches="tight")
```
